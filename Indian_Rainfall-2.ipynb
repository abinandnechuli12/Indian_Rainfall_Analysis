### Amount Of Rainfall

```python
df.groupby('subdivision')['JUN', 'JUL', 'AUG', 'SEP'].sum().plot(kind = 'bar', figsize = (15, 4), title='RainFall', xlabel = 'States', ylabel = 'Amount');
```

### Amount Of Rainfall in different years

```python
df.groupby('YEAR')['JUN-SEP'].sum().plot(kind = 'barh', figsize = (15, 19), title='RainFall', xlabel = 'States', ylabel = 'Amount');
```
